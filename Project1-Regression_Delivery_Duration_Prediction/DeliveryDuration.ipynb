```python
df2 = df2.dropna(subset=['actual_delivery_time'])
```

Convert datatime types

```python
df2['created_at'] = pd.to_timedelta(df2['created_at'])
df2['actual_delivery_time'] = pd.to_timedelta(df2['actual_delivery_time'])
```

# Create new feature for Target (Y)

```python
df2['delivery_duration'] = df2['actual_delivery_time'] - df2['created_at']
df2['delivery_duration'] = df2['delivery_duration'].dt.total_seconds()
```

```python
df2['created_at'] = df2['created_at'].dt.total_seconds()
df2['actual_delivery_time'] = df2['actual_delivery_time'].dt.total_seconds()
```

```python
    #   df2['created_hour'] = df2['created_at'].dt.hour
     #  df2['created_dayofweek'] = df2['created_at'].dt.dayofweek
     #  df2['created_day'] = df2['created_at'].dt.day
```

```python
df2['free_dashers'] = df2['total_onshift_dashers'] - df2['total_busy_dashers']
df2['free_dashers'] = df2['free_dashers'].clip(lower=0)
```

```python
df2 = df2[df2['delivery_duration'] > 0]
```

# Separate features (X) and target (Y) 

```python
X = df2.drop(columns=['delivery_duration','created_at','actual_delivery_time'])
Y = df2[['delivery_duration']]
```

# Data Preprocessing and Data Cleaning

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

# Data preprocessing and data cleaning by pipelines

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder())
    
    
```

```python
con_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler() )
```

```python
pre = ColumnTransformer([
    ('cat', cat_pipe,cat),
    ('con', con_pipe,con)
]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
```

```python
X_pre.head(5)
```

```python
new_cols = []
for i in X_pre:
    c = i.split('__')[-1]
    new_cols.append(c)
print(new_cols)
```

```python
X_pre.columns
```

```python
X_pre.columns = new_cols
X_pre.columns
```

```python
X_pre
```

# Feature Selection: Forward Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
base_model = LinearRegression()
for_sel = SequentialFeatureSelector(base_model,direction='forward',n_features_to_select=10)
for_sel.fit(X_pre,Y)
```

```python
for_sel.get_feature_names_out()
```

```python
imp_cols = for_sel.get_feature_names_out()
imp_cols
```

```python


len(imp_cols)

```

```python


sel_cols = []
for i in imp_cols:
    cols = i.split('__')[-1]
    sel_cols.append(cols)
sel_cols

```

```python
X_sel = X[sel_cols]
X_sel.head()
```

# 2nd step Data Preprocessing

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X_sel_cat = list(X_sel.columns[X_sel.dtypes=='object'])
X_sel_con = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
num_pipe2 = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler()
)
```

```python
cat_pipe2 = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore',sparse_output=False)
)
```

```python


pre2 = ColumnTransformer([
    ('cat',cat_pipe2,X_sel_cat),
    ('con',num_pipe2,X_sel_con)
]).set_output(transform='pandas')

```

```python


pre2

```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre,Y,train_size = 0.8,test_size = 0.2,random_state=21)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

# Build the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
ypreds = model.predict(xtest)
ypreds
```

# Evaluate the model
    Mean Sqaured Error
    Mean Absolute Error
    RMSE
    R2 score- Coeff of determination``

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse = mean_squared_error(ytest, ypreds)
mae = mean_absolute_error(ytest,ypreds)
rmse = mse**(1/2)
r2 = r2_score(ytest, ypreds)

print(f"MSE: {mse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R2 score: {r2*100:.2f}%")
```
